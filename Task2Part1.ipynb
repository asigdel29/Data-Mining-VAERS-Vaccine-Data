```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv('C://Users//sigde//cleandata.csv')
```

```python
final_df =df.groupby(["VAX_NAME", "VAX_TYPE","BIRTH_DEFECT","DIED","ALLERGIES"], as_index=False)["DIED"].count()
final_df
```

```python
main_df=pd.get_dummies(final_df,columns=["VAX_NAME", "VAX_TYPE","BIRTH_DEFECT","DIED","ALLERGIES"])
```

```python
frequent_itemsets = apriori(main_df,min_support=0.5,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets=fpgrowth(main_df,min_support=.05, use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules
```

```python
rules["antecedent_len"]=rules["antecedents"].apply(lambda x:len(x))
rules
```
