```python
df.drop(['SYMPTOM_TEXT'],axis=1,inplace=True)
df.drop(['ER_VISIT'],axis=1,inplace=True)
df.drop(['X_STAY'],axis=1,inplace=True)
df.drop(['LAB_DATA'],axis=1,inplace=True)
df.drop(['SPLTTYPE'],axis=1,inplace=True)
df.drop(['FORM_VERS'],axis=1,inplace=True)
df.drop(['VAX_LOT'],axis=1,inplace=True)
df.drop(['VAX_SITE'],axis=1,inplace=True)
df.drop(['PRIOR_VAX'],axis=1,inplace=True)
```
